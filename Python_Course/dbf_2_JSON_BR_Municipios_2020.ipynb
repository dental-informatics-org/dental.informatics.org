```python
import pandas as pd
from simpledbf import Dbf5
import json
import numpy


dbfPath = 'BR_Municipios_2020.dbf'

# Create a 'pandas.core.frame.DataFrame' from the DBF
df = Dbf5(dbfPath).to_dataframe()
# print (type(df)) # <class 'pandas.core.frame.DataFrame'>
# print (df)
# print ('-----') 

# Get an array of values from the 'pandas.core.frame.DataFrame' object
dfValues = df.values
# print (type(dfValues)) #<class 'numpy.ndarray'>
# print (dfValues)
# print ('-----') 

# Convert the 'numpy.ndarray' into a <class 'list'>
lists = dfValues.tolist()
# print (type(lists)) # <class 'list'>
# print (lists)
# print ('-----')

# Traverse the <class 'list'> and recreate a new one
GIS_list = [{'CD_MUN': x[0], 'NM_MUN': x[1], 'SIGLA_UF': x[2]} for i, x in enumerate(lists)]
# print (type(GIS_list)) # <class 'list'>
# print(GIS_list) # [{'CD_MUN': '1100015', 'NM_MUN': "Espigão D'Oeste", 'SIGLA_UF': 'RO'}, {'CD_MUN': '1100023',
# print ('-----')

# Convert the GIS_list into GIS_br_JSON and solve Encode issues
GIS_br_JSON = json.dumps(GIS_list).encode('utf-8')
# GIS_br_JSON = json.dumps(GIS_list)
print (type(GIS_br_JSON)) # <class 'str'>
print (GIS_br_JSON)
print ('-----')

# Save the JSON in a file
# with open('BR_Municipios_2020.json', 'a') as testwritefile:
#     testwritefile.write(GIS_br_JSON)
# print ('JSON exported')

# ======================================
# Faster way after Quang Hoang comments
dfDic = df.to_dict(orient='records')
# print (type(dfDic)) # <class 'list'>
# print (dfDic)
# print ('-----') 

GIS_br_JSON2 = json.dumps(dfDic)
# print (type(GIS_br_JSON2)) # <class 'str'>
# print (GIS_br_JSON2)
# print ('-----')

# Save the JSON in a file
# with open('BR_Municipios_2020.json', 'a') as testwritefile:
#     testwritefile.write(GIS_br_JSON2)
# print ('JSON exported')
```
